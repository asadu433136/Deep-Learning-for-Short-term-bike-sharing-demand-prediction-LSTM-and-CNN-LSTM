```python
import numpy as np
import pandas as pd
from datetime import datetime, timedelta
import seaborn as sns
```

```python
dataframe = pd.read_csv("database.csv", low_memory=False)

dataframe[['departure','return']] =  dataframe[['departure','return']].apply(pd.to_datetime, format='%Y-%m-%d %H:%M:%S.%f')

dataframe = dataframe.rename(columns={'distance (m)': 'distance',
                                     'duration (sec.)': 'duration',
                                     'avg_speed (km/h)':'speed',
                                      'Air temperature (degC)':'temperature',
                                     })
dataframe = dataframe[ 
                    (50 < dataframe['distance']) & (dataframe['distance'] < 10000) &
                    (120 < dataframe['duration']) & (dataframe['duration'] <  18000) &
                    (-20 < dataframe['temperature']) & (dataframe['temperature'] < 50)        
                    ].reset_index()
```

```python
dataframe = dataframe.sort_values("departure").reset_index(drop=True).drop('index', 1)
dataframe
```

```python
dataframe['timestamp'] = dataframe.apply(lambda row: row.departure - timedelta(minutes=(row.departure.minute%60)) - timedelta(seconds=row.departure.second), axis = 1)
dataframe
```

## Time function

```python
def groupby_count(df):
    unq, t = np.unique(df.departure_name, return_inverse=1)
    time = df.timestamp.values
    sidx = np.lexsort([t,time])

    ts = t[sidx]
    times = time[sidx]

    m0 = (times[1:] != times[:-1]) | (ts[1:] != ts[:-1])
    m = np.concatenate(([True], m0, [True]))
    time_out = times[m[:-1]]
    t_out = unq[ts[m[:-1]]]
    x_out = np.diff(np.flatnonzero(m)+1)

    out_ar = np.column_stack((time_out, t_out, x_out))
    return pd.DataFrame(out_ar, columns = ['timestamp','departure_name','pickup'])
```

```python
data_matrix = groupby_count(dataframe)
```

```python
data_matrix['timestamp'] = data_matrix['timestamp'].apply(pd.to_datetime, format='%Y-%m-%d %H:%M:%S.%f')
data_matrix
```

```python
data_matrix2 = pd.pivot_table(data_matrix, values='pickup', index=['timestamp'],
                    columns=['departure_name'], aggfunc=np.sum, fill_value=0)
data_matrix2
```

```python
# Calculate start and end time period: 
start_time = dataframe["departure"].min()
start_time = start_time - timedelta(minutes=(start_time.minute%60))
end_time = dataframe["departure"].max().replace(second=0)
end_time = end_time + timedelta(minutes=60 - (end_time.minute%60))

data_matrix_time = pd.DataFrame()
time_array = []
time = start_time

while time <= end_time:
    time_array.append(time)
    time =time + timedelta(minutes=60)
    
data_matrix_time["timestamp"] = time_array
```

```python
data = data_matrix_time.merge(data_matrix2, on='timestamp', how = 'left').fillna(0)
data.iloc[:,1:] = data.iloc[:,1:].astype('int')
```
